```python
import pandas as pd
import os
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

#Load dataset
path = r"C:\Users\aekmo\Downloads\Data_Pembelajaran_Mesin_DP\Data UTS\Mall_Customers.csv"
df = pd.read_csv(path)

#Hapus data kosong (jika ada)
df.dropna(inplace=True)

#Hapus duplikat (jika ada)
df.drop_duplicates(inplace=True)

#Encoding kolom Gender (Male = 1, Female = 0)
le = LabelEncoder()
df["Gender"] = le.fit_transform(df["Gender"])

#Normalisasi hanya kolom numerik
numeric_cols = df.select_dtypes(include=['number']).columns
scaler = MinMaxScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

#Tampilkan hasil normalisasi
print("Data setelah Normalisasi:")
print(df.head())

#Cek kembali apakah masih ada missing values
print("\nCek Missing Values:")
print(df.isnull().sum()[df.isnull().sum() > 0])

#Cek jumlah duplikat setelah dibersihkan
print(f"\nJumlah data duplikat: {df.duplicated().sum()}")

#Simpan hasil ke file CSV
save_path = r"C:\Users\aekmo\Downloads\Data_Pembelajaran_Mesin_DP\Data UTS\Mall_Customers_Normalized_Final.csv"
save_folder = os.path.dirname(save_path)

if not os.path.exists(save_folder):
    os.makedirs(save_folder)

df.to_csv(save_path, index=False)
print(f"\nDataset yang sudah dibersihkan dan dinormalisasi disimpan sebagai: '{save_path}'")
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

#Load dataset hasil preprocessing
path = r"C:\Users\aekmo\Downloads\Data_Pembelajaran_Mesin_DP\Data UTS\Mall_Customers_Normalized_Final.csv"
df = pd.read_csv(path)

#Terapkan KMeans Clustering
k = 5  #jumlah klaster yang diinginkan
kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(df)

#Tampilkan informasi awal
print("========== HASIL CLUSTERING DENGAN KMEANS ==========\n")
print(f"Jumlah klaster yang digunakan: {k}\n")
print("Contoh 10 data pertama setelah diklasterisasi:\n")
print(df.head(10))

#Tampilkan jumlah data per klaster
print("\nJumlah anggota tiap klaster:")
print(df['Cluster'].value_counts().sort_index())

#Komentar analisis hasil clustering
print("1. KMeans membagi data menjadi 5 kelompok (klaster) berdasarkan kemiripan fitur seperti usia, penghasilan, dan skor belanja.")
print("2. Setiap pelanggan dikategorikan ke dalam salah satu dari 5 kelompok.")
print("3. Jumlah anggota tiap klaster menunjukkan seberapa besar kelompok tersebut.")
print("4. Metode ini cocok jika kita ingin memahami segmentasi pelanggan untuk strategi pemasaran yang berbeda.")
print("5. Kelebihan KMeans: cepat, efisien, mudah dipahami.")
print("6. Kelemahan: Harus menentukan jumlah klaster di awal dan sensitif terhadap outlier.\n")

#PCA untuk reduksi dimensi ke 2D (agar bisa divisualisasikan)
pca = PCA(n_components=2)
principal_components = pca.fit_transform(df.drop("Cluster", axis=1))
df_pca = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
df_pca['Cluster'] = df['Cluster']

#Visualisasi hasil klaster
plt.figure(figsize=(8,6))
for i in range(k):
    cluster_data = df_pca[df_pca['Cluster'] == i]
    plt.scatter(cluster_data['PC1'], cluster_data['PC2'], label=f'Cluster {i}')

plt.title('Visualisasi Klaster KMeans dengan PCA (2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Komentar visualisasi
print("PENJELASAN VISUALISASI:")
print("Plot di atas menunjukkan representasi 2D dari data pelanggan yang telah diklaster menggunakan PCA.")
print("Setiap warna mewakili satu klaster.")
print("Kita bisa melihat pemisahan antar kelompok, yang menunjukkan bahwa pelanggan memang memiliki pola belanja yang berbeda.")
print("Visual ini sangat membantu untuk memahami segmentasi pasar secara visual.\n")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch

#Load dataset
path = r"C:\Users\aekmo\Downloads\Data_Pembelajaran_Mesin_DP\Data UTS\Mall_Customers_Normalized_Final.csv"
df = pd.read_csv(path)

#Gunakan hanya kolom numerik
numeric_df = df.select_dtypes(include=['float64', 'int64'])

#Plot dendrogram
plt.figure(figsize=(10, 6))
plt.title("Dendrogram untuk Menentukan Jumlah Klaster")
dendrogram = sch.dendrogram(sch.linkage(numeric_df, method='ward'))
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")
plt.tight_layout()
plt.show()

#Komentar dendrogram
print("\nPENJELASAN DAN KOMENTAR DENDROGRAM (Hierarchical Clustering):")
print("1. Dendrogram adalah representasi visual dari proses penggabungan klaster secara hierarkis.")
print("2. Setiap garis horizontal menunjukkan penggabungan dua klaster.")
print("3. Panjang garis menggambarkan perbedaan antar klaster (semakin panjang, semakin berbeda).")
print("4. Untuk memilih jumlah klaster, potong dendrogram secara horizontal di titik tertinggi yang tidak memotong garis horizontal lainnya.")
print("5. Hierarchical cocok untuk eksplorasi awal, tapi kurang efisien untuk dataset besar.")
```

```python
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load data hasil normalisasi
path = r"C:\Users\aekmo\Downloads\Data_Pembelajaran_Mesin_DP\Data UTS\Mall_Customers_Normalized_Final.csv"
df = pd.read_csv(path)

# ====================================
# DBSCAN Clustering
# ====================================
# eps = 0.3 : Jarak maksimum untuk menyatukan titik dalam satu klaster
# min_samples = 5 : Minimal jumlah titik di sekeliling (radius eps) agar titik dianggap sebagai "core point"
dbscan = DBSCAN(eps=0.3, min_samples=5)
df['Cluster_DBSCAN'] = dbscan.fit_predict(df)

# ====================================
# Informasi Hasil Klaster
# ====================================
print("HASIL CLUSTERING DENGAN DBSCAN\n")
print("Nilai klaster unik (termasuk noise -> label = -1):")
print(df['Cluster_DBSCAN'].value_counts().sort_index())

# Komentar hasil clustering
print("\nPenjelasan Algoritma DBSCAN:")
print("1. DBSCAN dapat membentuk klaster berdasarkan kepadatan data, tanpa perlu menentukan jumlah klaster di awal.")
print("2. Titik-titik yang tidak cukup padat atau terlalu jauh dari kelompok lainnya diberi label -1, artinya dianggap sebagai 'noise'.")
print("3. Cocok untuk data yang memiliki bentuk klaster tidak teratur dan mengandung outlier.")
print("4. Hasil di atas menunjukkan berapa banyak data yang masuk ke masing-masing klaster, dan berapa yang dianggap noise.")
print("5. Nilai parameter eps dan min_samples sangat mempengaruhi hasil. Jika terlalu kecil, banyak data dianggap noise.\n")

# ====================================
# Visualisasi dengan PCA
# ====================================
# PCA mengubah data berdimensi tinggi menjadi 2 dimensi agar bisa digambar
pca = PCA(n_components=2)
principal_components = pca.fit_transform(df.drop("Cluster_DBSCAN", axis=1))

df_pca = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
df_pca['Cluster'] = df['Cluster_DBSCAN']

plt.figure(figsize=(8,6))

for label in sorted(df_pca['Cluster'].unique()):
    cluster_data = df_pca[df_pca['Cluster'] == label]
    label_name = f"Cluster {label}" if label != -1 else "Noise"
    plt.scatter(cluster_data['PC1'], cluster_data['PC2'], label=label_name)

plt.title('Visualisasi Klaster DBSCAN dengan PCA (2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from scipy.cluster.hierarchy import linkage, fcluster, dendrogram

# Load dataset yang sudah dinormalisasi
path = r"C:\Users\aekmo\Downloads\Data_Pembelajaran_Mesin_DP\Data UTS\Mall_Customers_Normalized_Final.csv"
df = pd.read_csv(path)

# Ambil data numerik
data = df.select_dtypes(include=['float64', 'int64'])

# Standardisasi ulang agar DBSCAN lebih stabil
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# PCA untuk visualisasi
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

# =====================
# 1. KMEANS
# =====================
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(scaled_data)

# =====================
# 2. HIERARCHICAL
# =====================
linked = linkage(scaled_data, method='ward')
hierarchical_labels = fcluster(linked, 5, criterion='maxclust') - 1  # Subtract 1 for consistency

# =====================
# 3. DBSCAN
# =====================
dbscan = DBSCAN(eps=0.8, min_samples=5)
dbscan_labels = dbscan.fit_predict(scaled_data)

# =====================
# Visualisasi Hasil
# =====================
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

axs[0].scatter(pca_data[:, 0], pca_data[:, 1], c=kmeans_labels, cmap='tab10')
axs[0].set_title("KMeans Clustering")

axs[1].scatter(pca_data[:, 0], pca_data[:, 1], c=hierarchical_labels, cmap='tab10')
axs[1].set_title("Hierarchical Clustering")

axs[2].scatter(pca_data[:, 0], pca_data[:, 1], c=dbscan_labels, cmap='tab10')
axs[2].set_title("DBSCAN Clustering")

for ax in axs:
    ax.set_xlabel("PC1")
    ax.set_ylabel("PC2")
    ax.grid(True)

plt.suptitle("Perbandingan Visual Hasil Clustering", fontsize=16)
plt.tight_layout()
plt.show()

# =====================
# Tabel Ringkasan
# =====================
def cluster_summary(labels, name):
    unique = np.unique(labels)
    summary = {
        'Algoritma': name,
        'Jumlah Klaster': len(unique[unique != -1]),  # Tidak termasuk noise
        'Jumlah Noise': np.sum(labels == -1)
    }
    return summary

summary_kmeans = cluster_summary(kmeans_labels, "KMeans")
summary_hierarchical = cluster_summary(hierarchical_labels, "Hierarchical")
summary_dbscan = cluster_summary(dbscan_labels, "DBSCAN")

# Gabungkan hasil ke dataframe
summary_df = pd.DataFrame([summary_kmeans, summary_hierarchical, summary_dbscan])
print("\nTABEL PERBANDINGAN ALGORITMA CLUSTERING:")
print(summary_df.to_string(index=False))
```
