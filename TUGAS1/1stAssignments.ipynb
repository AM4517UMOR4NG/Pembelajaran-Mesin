```python
import matplotlib.pyplot as plt


column_1 = 'GA'  # Goals Against
column_2 = 'PSxG'  # Post-Shot Expected Goals

# Plot jumlah Goals Against dan PSxG per pemain
plt.figure(figsize=(12,6))
plt.bar(df['Player'], df[column_1], label=column_1, color='blue')
plt.bar(df['Player'], df[column_2], label=column_2, color='red', alpha=0.7)
plt.xticks(rotation=45)
plt.xlabel("Player")
plt.ylabel("Count")
plt.title(f"{column_1} vs {column_2} by Goalkeepers")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

#Plot Goals Against vs Post-Shot Expected Goals
plt.figure(figsize=(12,6))
plt.bar(df['Player'], df['GA'], label="Goals Against", color='blue')
plt.bar(df['Player'], df['PSxG'], label="Post-Shot Expected Goals", color='red', alpha=0.7)
plt.xticks(rotation=45)
plt.xlabel("Player")
plt.ylabel("Count")
plt.title("Goals Against vs Post-Shot Expected Goals by Goalkeepers")
plt.legend()
plt.show()
```

```python
# Plot Attacks vs Completion Percentage
plt.figure(figsize=(12,6))
plt.bar(df['Player'], df['Att'], label="Total Attempts", color='blue')
plt.bar(df['Player'], df['Cmp%'], label="Completion Percentage", color='green', alpha=0.7)
plt.xticks(rotation=45)
plt.xlabel("Player")
plt.ylabel("Count / Percentage")
plt.title("Total Attempts vs Completion Percentage by Goalkeepers")
plt.legend()
plt.show()
```
